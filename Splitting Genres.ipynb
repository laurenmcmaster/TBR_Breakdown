# Splitting Genres

Now I needed to create some extra tables to help with visualization. I needed to separate the genre lists into individual counts of each genre, so, first, I loaded my clean dataset into another Jupyter Notebook. Next, I split the genre list into individual columns, and used a for loop to iterate through columns and count each time a genre occurred. 

```python
import pandas as pd

data = pd.read_excel(r"C:\Users\lemcm\OneDrive\Goodreads Project Files\goodreads_data.xlsx")
data
```

```python
# splitting the genres into separate columns
genre_columns = data['Genres'].str.split(', ', expand=True)

# creating an empty dictionary to store genre counts (will become a dataframe)
genre_counts = {}

# iterating through the columns and count genre occurrences
for column in genre_columns.columns:
    counts = genre_columns[column].value_counts()
    for genre, count in counts.items():
        if genre in genre_counts:
            genre_counts[genre] += count
        else:
            genre_counts[genre] = count

# creating a dataframe from the genre counts
genre_counts_df = pd.DataFrame(list(genre_counts.items()), columns=['genre', 'genre_count'])

# printing dataframe
genre_counts_df
```

```python
# exporting the data to a csv
genre_counts_df.to_csv('genre_counts.csv', index=False)
```
